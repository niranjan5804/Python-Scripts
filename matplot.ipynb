```python
import numpy as np
import pandas as pd
from scipy.stats import norm
x=np.arange(-3,3,0.1)
y=norm.pdf(x)
df=pd.DataFrame({"x":x,"y":y})
df.head()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
df.plot(kind="scatter",x="x",y="y",c="blue")
```

```python
#Multiple plots
plt.scatter(x,y)
plt.plot(x,norm.pdf(x,1,0.5))
plt.show()
```

```python
#save the graph
plt.scatter(x,y,c="c")
plt.plot(x,norm.pdf(x,1,0.5),c="g")
plt.savefig(r"C:\Users\niranjan gowda ks\OneDrive\Pictures\plt.png")
```
