```python
import numpy as np
import pandas as pd
from scipy.stats import norm
x=np.arange(-3,3,0.1)
y=norm.pdf(x)
df=pd.DataFrame({"x":x,"y":y})
df.head()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
df.plot(kind="scatter",x="x",y="y",c="blue")
```

```python
#Multiple plots
plt.scatter(x,y)
plt.plot(x,norm.pdf(x,1,0.5))
plt.show()
```

```python
#save the graph
plt.scatter(x,y,c="c")
plt.plot(x,norm.pdf(x,1,0.5),c="g")
plt.savefig(r"C:\Users\niranjan gowda ks\OneDrive\Pictures\plt.png")
```

```python
#Customization of plots
```

```python
plt.scatter(x,y,c='c')
plt.plot(x,norm.pdf(x,1,0.5),c='g')
plt.xlim([-4,4]) #custom of axis values (x range from -4 to 4)
plt.ylim([0,1.2])
```

```python
plt.scatter(x,y,c='c')
plt.plot(x,norm.pdf(x,1,0.5),c='r')
plt.xticks([-3,0,1,3]) # only visible specified values
plt.yticks([0.4,0.8,0])
plt.xticks(rotation=90)# rotates the values
```

```python
plt.scatter(x,y,c='c')
plt.plot(x,norm.pdf(x,1,0.5),c='r')
plt.grid(linestyle="--",linewidth=0.7,axis="both",color="b")# linestyle= -,--,:,none,'',solid,dashed,dashdot,dotten
```

```python
plt.plot(x,norm.pdf(x,1,0.5),c='g',label="line plot")
plt.xlabel('event1',size=15,c='b',fontname='Times New Roman')
plt.ylabel('event2',size=15,c='r',fontname='Times New Roman')
```

```python
#legend
plt.scatter(x,y,c='c',label='scatter')
plt.plot(x,norm.pdf(x,1,0.5),c='r',label='line')
plt.grid(linestyle="--",linewidth=0.7,axis="both",color="b")
plt.legend(loc=2)#it prints the label contents specified in plots comand
#loc=1,2,3,4(4sides),center,center left_right_etc,upper left_right_etc,...
```

```python
#highliting the plots 
plt.annotate('this is plot',xy=(0,0.4),xytest=(-2,9,0.9),arrowprops=dict(facecolor="black"))
plt.show()
```
